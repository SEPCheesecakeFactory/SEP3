# Normalization of risk

Two numbers a and b, where both are 1-5, should be normalized in a way that the result is between 1 and 5 inclusive and should take into account the product of the two numbers.

```python
def normalize_risk(a, b):
    return (a - 1) * (b - 1) / 16 * 4 + 1
```

```python
print(normalize_risk(1, 1))
print(normalize_risk(2, 3))
print(normalize_risk(3, 3))
print(normalize_risk(5, 5))
```
